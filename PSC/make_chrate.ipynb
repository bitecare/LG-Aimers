## Import

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from tqdm.auto import tqdm
import seaborn as sns
```

## Data Load

```python
train_data = pd.read_csv("../DAT/train.csv")
day = pd.read_csv("../DAT/day_info.csv")
sales_d = pd.read_csv("../DAT/sales.csv")
```

## Create Data

```python
# 변동율
volt = pd.DataFrame(index = range(15890), columns = range(465)) # 데이터프레임 생성
volt.columns = train_data.columns
volt.iloc[:, :6] = train_data.iloc[:, :6]
volt[volt.columns[6]] = 0
```

```python
for i in tqdm(range(7, train_data.shape[1])): # 판매량 변동 계산
    volt[volt.columns[i]] = train_data.iloc[:, i] - train_data.iloc[:, i-1]
```

```python
for col in tqdm(range(7, 465)): # 판매량 변동이 0인 경우는 0, 아닌 경우는 역수 계산
    volt[volt.columns[col]] = volt[volt.columns[col]].apply(lambda x: 0 if x == 0 else 1 / x)
```

```python
# 기존 판매량 반영 변동율
volt1 = pd.DataFrame(index = range(15890), columns = range(465)) # 데이터 프레임 생성
volt1.columns = train_data.columns
volt1.iloc[:, :6] = train_data.iloc[:, :6]
volt1[volt1.columns[6]] = 0
for i in tqdm(range(7, train_data.shape[1])):
    volt1[volt1.columns[i]] = round((train_data.iloc[:, i] - train_data.iloc[:, i-1])/(train_data.iloc[:, i-1]+train_data.iloc[:, i]+1e-10), 5) # 전일 데이터 대비 판매량 변동을 전일과 당일 판매량 합으로 나눈 변수 생성
```

## Save

```python
volt.to_csv("../DAT/chrate.csv", index=False)
```

```python
volt1.to_csv("../DAT/chrate-base.csv", index=False)
```

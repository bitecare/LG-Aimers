## Load Result data

```python
# RandomForest
model_rf = pd.read_csv("../OUT/RFResult.csv")

# XGBoost
model_xgb = pd.read_csv("../OUT/XGBResult.csv")

#ExtraTree
model_xtr = pd.read_csv("../OUT/ExtraTreeResult.csv")

# LSTM
model_lstm = pd.read_csv("../OUT/LSTMResult.csv")
```

## ensemble!

```python
model_ensemble = model_rf + model_xgb + model_xtr+model_lstm
model_ensemble = model_ensemble/4
model_ensemble = np.round(model_ensemble,0).astype(int)
```

```python
modify = model_ensemble.iloc[:, 1:]
modify[modify<1] = 1 # 최저 판매량 값으로 대치
model_ensemble.iloc[:,1:] = modify
```

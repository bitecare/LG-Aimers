## make data

```python
# 시작 날짜와 종료 날짜 설정
start_date = '2022-01-01'
end_date = '2023-04-24'
# 시작 날짜부터 종료 날짜까지의 날짜 범위 생성
date_range = pd.date_range(start=start_date, end=end_date, freq='D')
# 빈 데이터프레임 생성
empty_df = pd.DataFrame(index=date_range)
empty_df['Year'] = empty_df.index.year
empty_df['Month'] = empty_df.index.month
empty_df['Day'] = empty_df.index.day
empty_df['Weekday'] = empty_df.index.weekday
weekday_names = ['월', '화', '수', '목', '금', '토', '일']
empty_df['Weekday_Name'] = empty_df['Weekday'].apply(lambda x: weekday_names[x])
empty_df['주말여부']='N'
empty_df.loc[(empty_df['Weekday_Name']=='토')|(empty_df['Weekday_Name']=='일'), '주말여부']='Y' # 주말
empty_df['공휴일여부']='N'
empty_df.loc[(empty_df['Month']==1)&(empty_df['Day']==1),'공휴일여부']='Y' # 신정
empty_df.loc[(empty_df['Month']==3)&(empty_df['Day']==1),'공휴일여부']='Y' # 삼일절
empty_df.loc[(empty_df['Month']==5)&(empty_df['Day']==5),'공휴일여부']='Y' # 어린이날
empty_df.loc[(empty_df['Month']==6)&(empty_df['Day']==6),'공휴일여부']='Y' # 현충일
empty_df.loc[(empty_df['Month']==8)&(empty_df['Day']==15),'공휴일여부']='Y' # 광복절
empty_df.loc[(empty_df['Month']==10)&(empty_df['Day']==3),'공휴일여부']='Y' # 개천절
empty_df.loc[(empty_df['Month']==10)&(empty_df['Day']==9),'공휴일여부']='Y' # 한글날
empty_df.loc[(empty_df['Month']==12)&(empty_df['Day']==25),'공휴일여부']='Y' # 성탄절
empty_df.loc[(empty_df['Year']==2023)&(empty_df['Month']==1)&((empty_df['Day']==21)|(empty_df['Day']==22)|(empty_df['Day']==23)|(empty_df['Day']==24)),'공휴일여부']='Y' # 2023 설연휴
empty_df.loc[(empty_df['Year']==2022)&(empty_df['Month']==1)&((empty_df['Day']==31)),'공휴일여부']='Y' # 2022 설연휴
empty_df.loc[(empty_df['Year']==2022)&(empty_df['Month']==2)&((empty_df['Day']==1)|(empty_df['Day']==2)),'공휴일여부']='Y' # 2022 설 연휴
empty_df.loc[(empty_df['Year']==2023)&(empty_df['Month']==9)&((empty_df['Day']==28)|(empty_df['Day']==29)|(empty_df['Day']==30)),'공휴일여부']='Y' # 2023 추석연휴
empty_df.loc[(empty_df['Year']==2023)&(empty_df['Month']==10)&((empty_df['Day']==1)|(empty_df['Day']==2)),'공휴일여부']='Y' # 2023 추석연휴
empty_df.loc[(empty_df['Year']==2022)&(empty_df['Month']==9)&((empty_df['Day']==9)|(empty_df['Day']==10)|(empty_df['Day']==11)|(empty_df['Day']==12)),'공휴일여부']='Y' # 2022 추석연휴
empty_df.loc[(empty_df['Year']==2022)&(empty_df['Month']==5)&((empty_df['Day']==8)),'공휴일여부']='Y' # 2022 석가탄신일
empty_df.loc[(empty_df['Year']==2022)&(empty_df['Month']==3)&((empty_df['Day']==9)),'공휴일여부']='Y' # 2022 대통령선거일
empty_df.loc[(empty_df['Year']==2022)&(empty_df['Month']==6)&((empty_df['Day']==1)),'공휴일여부']='Y' # 2022 지방선거일
empty_df.loc[(empty_df['Year']==2022)&(empty_df['Month']==10)&((empty_df['Day']==10)),'공휴일여부']='Y' # 2022 한글날 대체 휴일
empty_df['휴일여부']='N'
empty_df.loc[(empty_df['주말여부']=='Y')|(empty_df['공휴일여부']=='Y'), '휴일여부']='Y' # 주말이거나 공휴일이면 휴일
```

## Save

```python
empty_df.to_csv('/content/drive/MyDrive/phase3/DAT/day_info.csv', index=True, encoding='utf-8-sig')
```
